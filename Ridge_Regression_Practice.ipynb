```python
import seaborn as sns
import pandas as pd
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins['species'].unique()
```

```python
import numpy as np
```

```python
species = penguins.species.values[:, np.newaxis]
# species

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse_output=False)

onehot = enc.fit_transform(species)
print(enc.get_feature_names_out())
# onehot[::25]
onehot_df = pd.DataFrame(onehot, columns=enc.get_feature_names_out(), index=penguins.index)
onehot_df.columns = onehot_df.columns.str.replace('x0_', 'species_')
# onehot_df.head()
penguins_with_onehot = pd.concat([penguins, onehot_df], axis=1)
penguins_with_onehot[::25]
```

```python
penguins['island'].unique()
```

```python
island = penguins.island.values[:, np.newaxis]
# island
# enc = OneHotEncoder(sparse_output=False)
onehot_island = enc.fit_transform(island)
# print(enc.get_feature_names_out())
onehot_island_df = pd.DataFrame(onehot_island, columns=enc.get_feature_names_out(), index=penguins.index)
# onehot_island_df
onehot_island_df.columns = onehot_island_df.columns.str.replace('x0_', 'island_')
# onehot_island_df.head()
penguins_with_onehot_island = pd.concat([penguins_with_onehot, onehot_island_df], axis=1)
penguins_with_onehot_island[::25]
```
